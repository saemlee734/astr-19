```python

import numpy as np
import matplotlib.pyplot as plt

```

```python

random_numbers = np.random.uniform(0, 1, 1000)

plt.hist(random_numbers, bins=100)

plt.xlabel('Value')
plt.ylabel('Frequency')

plt.title('Histogram of 1000 Random Numbers')
```

```python
# Saving the figure as a PDF
plt.savefig('random_numbers_histogram.pdf')

plt.show()
```
