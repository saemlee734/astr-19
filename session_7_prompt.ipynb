```python
import matplotlib as plt

from matplotlib.pyplot import plot, xlabel, ylabel, title, show

import numpy as np
```

```python
x = np.linspace(0, 1, 100)
```

```python
def exp_function(x):
    return np.exp(x)
```

```python
y = exp_function(x)

```

```python
plot(x, y)
xlabel('Time [milliseconds]')
ylabel('Awesomeness')
title('Awesomeness over Time')
show()

```
