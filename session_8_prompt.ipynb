```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def sin_function(x):
    return np.sin(x)

def cos_function(x):
    return np.cos(x)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot sin(x)
axs[0].plot(x, sin_function(x))
axs[0].set_title('sin(x)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('sin(x)')

# Plot cos(x)
axs[1].plot(x, cos_function(x))
axs[1].set_title('cos(x)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('cos(x)')

plt.tight_layout()

# Save figure as PDF
plt.savefig('sin_cos_plot.pdf')

plt.show()
```
